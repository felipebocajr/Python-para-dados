## Manipulando os dados

Criando colunas numéricas

```python
#atribui a uma variável a leitura do link com os dados, e então retorna o cabeçalho
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep=';')
dados.head()
```

```python
#cria uma nova coluna 'Valor_por_mes' que é a soma do valor + condominio
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
dados.head()
```

```python
#cria uma nova coluna 'Valor_por_ano' que é a soma do valor por ano + IPTU
dados['Valor_por_ano'] = dados['Valor_por_mes'] + dados['IPTU']
dados.head()
```

**Criando colunas categóricas**

```python
#cria uma nova coluna 'Descrição' que é a soma do texto de tipo de imóvel + em + nome do bairro
dados['Descrição'] = dados['Tipo'] + ' em ' + dados['Bairro']
dados.head()
```

```python
#sobrescreve coluna 'Descrição' com a soma do tipo de imóvel, seu bairro, número de quartos e vagas disponíveis
dados['Descrição'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + dados['Quartos'].astype(str) + ' quarto(s) ' + ' e ' + dados['Vagas'].astype(str) + ' vaga(s)'
dados.head()
```

```python
#cria uma nova coluna 'possui_suite' utiliza a função apply, em seguida é dado uma função lambda que faz um mapeamento em todas as linhas
#e então passa a condição: se a coluna de suite for 0, a coluna 'possui_suite' retorna 'não', caso contrário retorna 'sim'
dados['Possui_suite'] = dados['Suites'].apply(lambda x: "sim" if x > 0 else "não")
dados.head()
```

```python
#salvando em um arquivo CSV, sem index e com separação ';'
dados.to_csv('dados_completos_dev.csv', index=False, sep=';')
```

**Desafios**

*1) Os alunos participaram de uma atividade extracurricular e ganharam pontos extras. Esses pontos extras correspondem a 40% da nota atual de cada um deles. Com base nisso, crie uma coluna chamada "Pontos_extras" que contenha os pontos extras de cada aluno, ou seja, 40% da nota atual deles.*

```python
#importa o link que contem os dados em CSV e atribui a uma variável
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
alunos = pd.read_csv(url, sep=',')
alunos.head()
```

```python
#cria uma nova coluna 'pontos_extras' e passa uma função lambda que atribui a essa coluna 40% da nota do aluno
alunos['Pontos_extras'] = alunos['Notas'].apply(lambda x: x*0.4)
alunos.head()
```

*2) Crie mais uma coluna, chamada "Notas_finais" que possua as notas de cada aluno somada com os pontos extras.*

```python
#cria uma nova coluna 'notas_finais' que soma as notas com os pontos extras
alunos['Notas_finais'] = alunos['Notas'] + alunos['Pontos_extras']
alunos.head()
```

*3) Como houve uma pontuação extra, alguns alunos que não tinham sido aprovados antes podem ter sido aprovados agora. Com base nisso, crie uma coluna chamada "Aprovado_final" com os seguintes valores:


True: caso o aluno esteja aprovado (nota final deve ser maior ou igual a 6);

False: caso o aluno esteja reprovado (nota final deve ser menor que 6).*

```python
#cria uma nova coluna 'aprovado' que agora se define a partir das notas finais, e não apenas das notas
alunos['Aprovado'] = alunos['Notas_finais'].apply(lambda x: True if x >= 6 else False)
alunos.head()
```

*4) Faça uma seleção e verifique quais alunos não tinham sido aprovados anteriormente, mas foram aprovados após a soma dos pontos extras.*

```python
#retorna apenas alunos com nota abaixo de 6 e nota final maior ou igual a 6
alunos.query('Notas < 6 & Notas_finais >= 6')
```
