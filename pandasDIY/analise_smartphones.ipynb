```python
import pandas as pd
```

```python
phones = pd.read_csv('smartphones.csv')
phones.query("`Final Price` > 300.00")
```

```python
phones.head()
```

```python
#criando uma nova coluna
phones['categoria'].apply(lambda x: "Flagship" if phones['RAM'] >=8 & phones['Final Price'] > 699 else 
                          "Mid-range" if phones['RAM'] <= 8 and phones['RAM'] >= 6 and phones['Final Price'] < 699 and phones['Final Price'] >= 299 else 
                          "Budget" if phones['RAM'] < 4 and phones['Final Price'] < 299 else "unknown")
```

```python
phones['categoria'] = phones.apply(
    lambda row: "Flagship" if (row['RAM'] >= 8) & (row['Final Price'] > 699) else 
                "Mid-range" if (row['RAM'] <= 8) & (row['RAM'] >= 6) & (row['Final Price'] < 699) & (row['Final Price'] >= 299) else 
                "Budget" if (row['RAM'] < 4) & (row['Final Price'] < 299) else 
                "unknown",
    axis=1
)
```
